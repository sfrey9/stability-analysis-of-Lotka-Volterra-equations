```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma

X, Y = np.meshgrid(np.arange(0, 8, .5), np.arange(0, 8, .5))
b=2
a=0.5
U = 100*X*(b-X-Y/(1+X));
V = 100*Y*(X/(1+X)-a*Y)
t = np.arange(0,8,0.001)
Z = (b-t)*(1+t)
W = t/(a*(1+t))

plt.figure(figsize=(6,4))
plt.title('Phase Space, a=0.5, b=2')
Q = plt.quiver(X, Y, U, V)
#qk = plt.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
#    coordinates='figure')
line1, = plt.plot(t,Z, label="x'=0")
line2, = plt.plot(t,W, label="y'=0")
axes = plt.gca()
axes.set_xlim([0,8])
axes.set_ylim([0,8])
plt.legend(handles=[line1,line2], loc=1)
axes.set_xlabel("x")
axes.set_ylabel("y")


plt.show()
```
