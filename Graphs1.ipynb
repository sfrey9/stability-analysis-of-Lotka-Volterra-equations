```python
import matplotlib.pyplot as plt
% matplotlib inline

def f(x,y):
    return x*(3.5-x-y/(1+x))
def g(x,y):
    return y*(x/(1+x)-0.1*y)

x = []
y = []

def sys(iv1, iv2, dt, time):
    x.append(iv1)
    y.append(iv2)
    for i in range(time):
        x.append(x[i] + (f(x[i],y[i])) * dt)
        y.append(y[i] + (g(x[i],y[i])) * dt)
    return x, y

sys(0.9, 0.1, 0.01, 10000)

#plot
fig = plt.figure(figsize=(15,5))
fig.subplots_adjust(wspace = 0.5, hspace = 0.3)
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.plot(x, 'r-', label='x')
ax1.plot(y, 'b-', label='y')

ax1.set_title("a=0.1, b=1, x0=0.9, y0=0.1")
ax1.set_xlabel("time")
ax1.grid()
ax1.legend(loc='best')

ax2.plot(x, y, color="blue")
ax2.set_xlabel("x")
ax2.set_ylabel("y")  
ax2.set_title("Phase space")
ax2.grid()
```
